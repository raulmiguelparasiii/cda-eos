```python
import json, zipfile, subprocess, sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

SCRIPT = Path("cda_eos_numeric_proof.py")
assert SCRIPT.exists(), "Run this in the folder containing cda_eos_numeric_proof.py"

runs = [
    ("curv_step_05", "5e-5", "5e-5"),
    ("curv_step_10", "1e-4", "1e-4"),
    ("curv_step_20", "2e-4", "2e-4"),
]

nT_R = 11
nrho_R = 9

for name, du, dv in runs:
    out = f"outputs_{name}"
    cmd = [
        sys.executable, str(SCRIPT),
        "--backend", "coolprop",
        "--enable-curvature",
        "--n-T-R", str(nT_R),
        "--n-rho-R", str(nrho_R),
        "--du-rel", du,
        "--dv-rel", dv,
        "--out", out,
    ]
    print("RUNNING:", " ".join(cmd))
    subprocess.run(cmd, check=True)

def load_run(name):
    d = Path(f"outputs_{name}")
    npz = np.load(d / "curvature_grid.npz")
    with open(d / "curvature_report.json") as f:
        rep = json.load(f)
    return d, npz, rep

loaded = [load_run(name) for name, _, _ in runs]

Ts = loaded[1][1]["Ts"]
rhos = loaded[1][1]["rhos"]

R_stack = []
valid_stack = []

for d, npz, rep in loaded:
    R = npz["R"]
    invalid = npz["invalid"].astype(bool)
    valid = np.isfinite(R) & (~invalid)
    R_stack.append(R)
    valid_stack.append(valid)

R_stack = np.stack(R_stack, axis=0)
valid_all = np.logical_and.reduce(valid_stack)

absR = np.abs(R_stack)
logR = np.log1p(absR)

logR_mean = np.nanmean(np.where(valid_all[None, :, :], logR, np.nan), axis=0)
logR_std = np.nanstd(np.where(valid_all[None, :, :], logR, np.nan), axis=0)

robust_mask = valid_all & (logR_std < 0.75)

median_R = np.nanmedian(np.where(valid_all[None, :, :], R_stack, np.nan), axis=0)
median_absR = np.abs(median_R)
finite_abs = median_absR[np.isfinite(median_absR) & robust_mask]

if finite_abs.size:
    R0 = float(np.nanmedian(finite_abs))
    if not np.isfinite(R0) or R0 <= 0:
        R0 = 1e-6
else:
    R0 = 1e-6

Reps = max(1e-6 * R0, 1e-30)
n_median = 1.0 - np.exp(-np.sqrt(median_R * median_R + Reps * Reps) / R0)
n_median = np.where(robust_mask, n_median, np.nan)

report = {
    "grid_shape": [int(len(Ts)), int(len(rhos))],
    "runs": [
        {"name": name, "du_rel": du, "dv_rel": dv, "report": rep}
        for (name, du, dv), (_, _, rep) in zip(runs, loaded)
    ],
    "valid_all_cells": int(np.sum(valid_all)),
    "robust_cells": int(np.sum(robust_mask)),
    "robust_fraction_of_grid": float(np.mean(robust_mask)),
    "robust_fraction_of_valid_all": float(np.sum(robust_mask) / max(np.sum(valid_all), 1)),
    "R0_from_robust_median_absR": R0,
    "R_epsilon": Reps,
    "median_R_min_robust": float(np.nanmin(median_R[robust_mask])) if np.any(robust_mask) else float("nan"),
    "median_R_max_robust": float(np.nanmax(median_R[robust_mask])) if np.any(robust_mask) else float("nan"),
    "median_R_median_robust": float(np.nanmedian(median_R[robust_mask])) if np.any(robust_mask) else float("nan"),
    "n_median_min_robust": float(np.nanmin(n_median[robust_mask])) if np.any(robust_mask) else float("nan"),
    "n_median_max_robust": float(np.nanmax(n_median[robust_mask])) if np.any(robust_mask) else float("nan"),
    "n_median_median_robust": float(np.nanmedian(n_median[robust_mask])) if np.any(robust_mask) else float("nan"),
}

outdir = Path("outputs_curvature_robustness")
outdir.mkdir(exist_ok=True)

with open(outdir / "curvature_robustness_report.json", "w") as f:
    json.dump(report, f, indent=2)

np.savez_compressed(
    outdir / "curvature_robustness_fields.npz",
    Ts=Ts,
    rhos=rhos,
    R_step_05=R_stack[0],
    R_step_10=R_stack[1],
    R_step_20=R_stack[2],
    valid_all=valid_all,
    robust_mask=robust_mask,
    median_R=median_R,
    n_median=n_median,
    logR_std=logR_std,
)

def heatmap(Z, title, fname, cbar_label=""):
    plt.figure(figsize=(7, 4.8))
    cmap = plt.cm.viridis.copy()
    cmap.set_bad(color="lightgray")
    plt.imshow(np.ma.masked_invalid(Z), origin="lower", aspect="auto",
               extent=[float(rhos[0]), float(rhos[-1]), float(Ts[0]), float(Ts[-1])],
               cmap=cmap)
    cb = plt.colorbar()
    if cbar_label:
        cb.set_label(cbar_label)
    plt.title(title)
    plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
    plt.ylabel("T [K]")
    plt.tight_layout()
    plt.savefig(outdir / fname, dpi=180)
    plt.close()

heatmap(np.where(robust_mask, 1.0, np.nan), "Robust curvature cells", "heatmap_robust_mask.png")
heatmap(np.where(robust_mask, median_R, np.nan), "Median R on robust cells", "heatmap_median_R_robust.png", "m^3/mol, convention-dependent")
heatmap(n_median, "Median network factor n on robust cells", "heatmap_n_median_robust.png")
heatmap(np.where(valid_all, logR_std, np.nan), "Step-size sensitivity: std(log(1+|R|))", "heatmap_logR_std.png")

print(json.dumps(report, indent=2))

zip_name = "curvature_robustness_bundle.zip"
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for folder in [*(Path(f"outputs_{name}") for name, _, _ in runs), outdir]:
        for p in folder.rglob("*"):
            z.write(p, p)

print("Created:", zip_name)
```

```python
# CDA-EOS next curvature pass: direct Ruppeiner metric in (T, v_m) coordinates.
#
# Use this in Colab in the same folder where cda_eos_numeric_proof.py exists.
# It avoids the noisy nested s(u,v) Hessian route.
#
# Metric used:
#   g_TT = c_v,m / (R T^2)
#   g_vv = - (dp/dv)_T / (R T)
# where v_m = molar volume and dp/dv is molar-volume derivative.
#
# Since CoolProp gives rho_mass derivatives:
#   rho_m = rho_mass / M
#   v_m = 1/rho_m = M/rho_mass
#   dp/dv_m = dp/drho_mass * d rho_mass/dv_m
#            = dp/drho_mass * ( - M / v_m^2 )
#
# Then compute scalar curvature from the 2D metric by finite differences on a grid.

import json, zipfile, sys, subprocess
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

!pip install -q CoolProp numpy matplotlib scipy

from CoolProp.CoolProp import PropsSI, PhaseSI

R_GAS = 8.31446261815324
M_WATER = 0.018015268

OUT = Path("outputs_metric_curvature")
OUT.mkdir(exist_ok=True)

# Keep conservative/small first.
T_min, T_max, nT = 250.0, 350.0, 61
rho_min, rho_max, nrho = 900.0, 1100.0, 51

Ts = np.linspace(T_min, T_max, nT)
rhos = np.linspace(rho_min, rho_max, nrho)

def is_bad_state(T, rho):
    try:
        ph = str(PhaseSI("T", float(T), "Dmass", float(rho), "Water")).lower()
        if ("two" in ph) or ("twophase" in ph) or ("gas" in ph) or ("vapor" in ph):
            return True
    except Exception:
        pass
    try:
        rhoL = float(PropsSI("Dmass", "T", float(T), "Q", 0, "Water"))
        if rho < rhoL - 1e-4*rhoL:
            return True
    except Exception:
        pass
    return False

# Fields in coordinates (T, v_m)
vms = M_WATER / rhos   # m^3/mol

gTT = np.full((nT, nrho), np.nan)
gvv = np.full((nT, nrho), np.nan)
invalid = np.zeros((nT, nrho), dtype=bool)

for i,T in enumerate(Ts):
    for j,rho in enumerate(rhos):
        try:
            if is_bad_state(T, rho):
                invalid[i,j] = True
                continue

            cv_mass = float(PropsSI("Cvmass", "T", float(T), "Dmass", float(rho), "Water"))
            cv_m = cv_mass * M_WATER

            dp_drho_mass = float(PropsSI("d(P)/d(Dmass)|T", "T", float(T), "Dmass", float(rho), "Water"))
            v_m = M_WATER / rho
            dp_dv_m = dp_drho_mass * (-M_WATER / (v_m*v_m))

            gTT[i,j] = cv_m / (R_GAS * T*T)
            gvv[i,j] = -dp_dv_m / (R_GAS * T)

            if not (np.isfinite(gTT[i,j]) and np.isfinite(gvv[i,j]) and gTT[i,j] > 0 and gvv[i,j] > 0):
                invalid[i,j] = True

        except Exception:
            invalid[i,j] = True

valid = (~invalid) & np.isfinite(gTT) & np.isfinite(gvv) & (gTT > 0) & (gvv > 0)

# Work on the full rectangular grid; mark R invalid where derivatives touch bad cells.
Rfield = np.full_like(gTT, np.nan, dtype=float)

# Coordinates: x0=T, x1=v_m.
# Metric diagonal: g00=A, g11=B.
A = gTT
B = gvv

# helper finite differences with numpy gradients
# Edge derivatives are lower quality; we will mask boundary.
dA_dT, dA_dv = np.gradient(A, Ts, vms, edge_order=2)
dB_dT, dB_dv = np.gradient(B, Ts, vms, edge_order=2)

# Christoffel symbols for diagonal metric.
# Gamma[k,i,j]
Gamma = np.full((2,2,2,nT,nrho), np.nan)
invA = 1.0/A
invB = 1.0/B

# Γ^T_TT = 0.5 A^-1 A_T
Gamma[0,0,0] = 0.5 * invA * dA_dT
# Γ^T_Tv = Γ^T_vT = 0.5 A^-1 A_v
Gamma[0,0,1] = Gamma[0,1,0] = 0.5 * invA * dA_dv
# Γ^T_vv = -0.5 A^-1 B_T
Gamma[0,1,1] = -0.5 * invA * dB_dT
# Γ^v_TT = -0.5 B^-1 A_v
Gamma[1,0,0] = -0.5 * invB * dA_dv
# Γ^v_Tv = Γ^v_vT = 0.5 B^-1 B_T
Gamma[1,0,1] = Gamma[1,1,0] = 0.5 * invB * dB_dT
# Γ^v_vv = 0.5 B^-1 B_v
Gamma[1,1,1] = 0.5 * invB * dB_dv

# Derivatives of Gamma wrt T and v
dG = {}
for k in range(2):
    for i in range(2):
        for j in range(2):
            dT, dv = np.gradient(Gamma[k,i,j], Ts, vms, edge_order=2)
            dG[(k,i,j,0)] = dT
            dG[(k,i,j,1)] = dv

Ric = np.full((2,2,nT,nrho), np.nan)
for i in range(2):
    for j in range(2):
        term = np.zeros((nT,nrho), dtype=float)
        # ∂_k Γ^k_ij
        term1 = sum(dG[(k,i,j,k)] for k in range(2))
        # ∂_j Γ^k_ik
        term2 = sum(dG[(k,i,k,j)] for k in range(2))
        term3 = np.zeros_like(term1)
        term4 = np.zeros_like(term1)
        for k in range(2):
            for l in range(2):
                term3 += Gamma[k,i,j] * Gamma[l,k,l]
                term4 += Gamma[l,i,k] * Gamma[k,j,l]
        Ric[i,j] = term1 - term2 + term3 - term4

Rfield = invA * Ric[0,0] + invB * Ric[1,1]

# Mask boundary and any cell adjacent to invalid cells because gradients need neighbors.
robust_valid = valid.copy()
robust_valid[0,:] = robust_valid[-1,:] = False
robust_valid[:,0] = robust_valid[:,-1] = False
for di in [-1,0,1]:
    for dj in [-1,0,1]:
        shifted_bad = np.zeros_like(invalid)
        if di >= 0:
            src_i = slice(0, nT-di)
            dst_i = slice(di, nT)
        else:
            src_i = slice(-di, nT)
            dst_i = slice(0, nT+di)
        if dj >= 0:
            src_j = slice(0, nrho-dj)
            dst_j = slice(dj, nrho)
        else:
            src_j = slice(-dj, nrho)
            dst_j = slice(0, nrho+dj)
        shifted_bad[dst_i, dst_j] = invalid[src_i, src_j]
        robust_valid &= ~shifted_bad

Rfield = np.where(robust_valid & np.isfinite(Rfield), Rfield, np.nan)

absR = np.abs(Rfield)
vals = absR[np.isfinite(absR)]
R0 = float(np.nanmedian(vals)) if vals.size else 1e-6
if not np.isfinite(R0) or R0 <= 0:
    R0 = 1e-6
Reps = max(1e-6 * R0, 1e-30)
nfield = 1.0 - np.exp(-np.sqrt(Rfield*Rfield + Reps*Reps)/R0)

report = {
    "method": "direct diagonal Ruppeiner metric in (T, v_m)",
    "grid_shape": [int(nT), int(nrho)],
    "valid_metric_cells": int(np.sum(valid)),
    "robust_R_cells": int(np.sum(np.isfinite(Rfield))),
    "frac_robust_R": float(np.mean(np.isfinite(Rfield))),
    "R_min": float(np.nanmin(Rfield)) if vals.size else float("nan"),
    "R_max": float(np.nanmax(Rfield)) if vals.size else float("nan"),
    "R_median": float(np.nanmedian(Rfield)) if vals.size else float("nan"),
    "absR_median_R0": R0,
    "R_epsilon": Reps,
    "n_min": float(np.nanmin(nfield)) if vals.size else float("nan"),
    "n_max": float(np.nanmax(nfield)) if vals.size else float("nan"),
    "n_median": float(np.nanmedian(nfield)) if vals.size else float("nan"),
}

with open(OUT / "metric_curvature_report.json", "w") as f:
    json.dump(report, f, indent=2)

np.savez_compressed(
    OUT / "metric_curvature_fields.npz",
    Ts=Ts,
    rho_mass=rhos,
    v_molar=vms,
    gTT=gTT,
    gvv=gvv,
    valid=valid,
    robust_valid=robust_valid,
    R=Rfield,
    n=nfield,
)

def heatmap(Z, title, fname, cbar_label=""):
    plt.figure(figsize=(8,5))
    cmap = plt.cm.viridis.copy()
    cmap.set_bad(color="lightgray")
    plt.imshow(np.ma.masked_invalid(Z), origin="lower", aspect="auto",
               extent=[float(rhos[0]), float(rhos[-1]), float(Ts[0]), float(Ts[-1])],
               cmap=cmap)
    cb = plt.colorbar()
    if cbar_label:
        cb.set_label(cbar_label)
    plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
    plt.ylabel("T [K]")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(OUT / fname, dpi=180)
    plt.close()

heatmap(gTT, "Metric component g_TT", "heatmap_gTT.png")
heatmap(gvv, "Metric component g_vv", "heatmap_gvv.png")
heatmap(Rfield, "Ruppeiner curvature R via direct metric", "heatmap_metric_R.png", "m^3/mol, convention-dependent")
heatmap(nfield, "Network factor n from direct-metric R", "heatmap_metric_n.png")
heatmap(np.where(robust_valid, 1.0, np.nan), "Robust valid cells for direct metric R", "heatmap_metric_valid_mask.png")

print(json.dumps(report, indent=2))

zip_name = "metric_curvature_bundle.zip"
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for p in OUT.rglob("*"):
        z.write(p, p)
print("Created:", zip_name)
```

```python
# CDA-EOS association correction sweep
# Run this after the base CoolProp run and the direct-metric curvature run.
# It uses:
# - base CoolProp fields
# - direct-metric n(T,rho)
# - SAFT-style a_assoc
# - pressure correction from the same free-energy term

import json, csv, zipfile, math
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

try:
    from scipy.interpolate import griddata
    from scipy.ndimage import gaussian_filter
except Exception:
    !pip install -q scipy
    from scipy.interpolate import griddata
    from scipy.ndimage import gaussian_filter

R_GAS = 8.31446261815324
M_WATER = 0.018015268
M_SITES = 4

OUT = Path("outputs_assoc_sweep")
OUT.mkdir(exist_ok=True)

def find_file(filename):
    hits = list(Path(".").rglob(filename))
    if not hits:
        raise FileNotFoundError(f"Could not find {filename}. Make sure outputs_base and outputs_metric_curvature exist.")
    return hits[0]

base_npz_path = find_file("base_fields_grid.npz")
metric_npz_path = find_file("metric_curvature_fields.npz")

print("Using base file:", base_npz_path)
print("Using metric curvature file:", metric_npz_path)

base = np.load(base_npz_path)
metric = np.load(metric_npz_path)

Ts = base["Ts"]
rho_mass = base["rhos"]
rho_mol = rho_mass / M_WATER

P_base = base["p"]
alpha_base = base["alpha_p"]
kappa_base = base["kappa_T"]
cp_base = base["cp"]
invalid_base = base["invalid"].astype(bool)

T_metric = metric["Ts"]
rho_metric = metric["rho_mass"]
n_metric = metric["n"]
R_metric = metric["R"]

valid_n = np.isfinite(n_metric)

TT, RR = np.meshgrid(Ts, rho_mass, indexing="ij")

points = []
vals = []

for i, T in enumerate(T_metric):
    for j, rho in enumerate(rho_metric):
        if valid_n[i, j]:
            points.append((float(T), float(rho)))
            vals.append(float(n_metric[i, j]))

points = np.array(points)
vals = np.array(vals)

if len(vals) < 3:
    raise RuntimeError("Not enough valid n(T,rho) points from metric curvature output.")

n_interp = griddata(points, vals, (TT, RR), method="linear")
n_near = griddata(points, vals, (TT, RR), method="nearest")
n_interp = np.where(np.isfinite(n_interp), n_interp, n_near)
n_interp = np.clip(n_interp, 0.0, 1.0)

# Smooth lightly so finite derivatives do not chase pixel-level interpolation noise.
n_grid = gaussian_filter(n_interp, sigma=1.5)
n_grid = np.clip(n_grid, 0.0, 1.0)

def tmd_from_alpha(alpha, invalid_mask):
    out = {}
    for j, rho in enumerate(rho_mass):
        a = alpha[:, j]
        bad = invalid_mask[:, j] | ~np.isfinite(a)

        for i in range(len(Ts) - 1):
            if bad[i] or bad[i + 1]:
                continue

            a0 = a[i]
            a1 = a[i + 1]

            if a0 == 0:
                out[float(rho)] = float(Ts[i])
                break

            if a0 * a1 < 0:
                T0 = Ts[i]
                T1 = Ts[i + 1]
                Tcross = T0 - a0 * (T1 - T0) / (a1 - a0)
                out[float(rho)] = float(Tcross)
                break

    return out

base_tmd = tmd_from_alpha(alpha_base, invalid_base)

def compute_trial(K0, epsK, alpha0, lam):
    """
    K0: effective bonding volume [m^3/mol]
    epsK: epsilon/R in kelvin, so exp(epsilon/(RT)) = exp(epsK/T)
    alpha0: baseline association factor
    lam: CDA coupling strength
    """
    Tmat, rhomat = np.meshgrid(Ts, rho_mol, indexing="ij")

    Delta0 = K0 * (np.exp(epsK / Tmat) - 1.0)
    Delta = Delta0 * (alpha0 + (1.0 - alpha0) * n_grid)

    A = rhomat * M_SITES * Delta

    # Stable small-A handling.
    X = np.where(
        A < 1e-12,
        1.0 - A,
        (-1.0 + np.sqrt(1.0 + 4.0 * A)) / (2.0 * A),
    )

    X = np.clip(X, 1e-12, 1.0)

    a_assoc = R_GAS * Tmat * M_SITES * (np.log(X) - X / 2.0 + 0.5)

    da_drho = np.gradient(a_assoc, rho_mol, axis=1, edge_order=2)
    p_assoc = (rho_mol[None, :] ** 2) * da_drho

    P_total = P_base + lam * p_assoc

    dP_dT = np.gradient(P_total, Ts, axis=0, edge_order=2)
    dP_drho_mol = np.gradient(P_total, rho_mol, axis=1, edge_order=2)

    alpha_total = (1.0 / rho_mol[None, :]) * dP_dT / dP_drho_mol
    kappa_total = 1.0 / (rho_mol[None, :] * dP_drho_mol)

    valid = (
        (~invalid_base)
        & np.isfinite(P_total)
        & np.isfinite(alpha_total)
        & np.isfinite(kappa_total)
        & np.isfinite(dP_drho_mol)
    )

    denom = max(int(np.sum(valid)), 1)

    bad_mech = ((dP_drho_mol <= 0.0) | (kappa_total <= 0.0)) & valid

    tmd_total = tmd_from_alpha(alpha_total, invalid_base | (~valid))

    common = sorted(set(base_tmd.keys()).intersection(tmd_total.keys()))
    shifts = [tmd_total[r] - base_tmd[r] for r in common]

    p_corr = lam * p_assoc

    result = {
        "K0_m3_per_mol": float(K0),
        "epsK_K": float(epsK),
        "alpha0": float(alpha0),
        "lambda": float(lam),
        "valid_cells": int(np.sum(valid)),
        "bad_mechanical_cells": int(np.sum(bad_mech)),
        "bad_mechanical_frac_valid": float(np.sum(bad_mech) / denom),
        "min_dP_drho_molar_valid": float(np.nanmin(dP_drho_mol[valid])) if np.any(valid) else float("nan"),
        "min_kappa_total_valid": float(np.nanmin(kappa_total[valid])) if np.any(valid) else float("nan"),
        "p_correction_median_MPa": float(np.nanmedian(p_corr[valid]) / 1e6) if np.any(valid) else float("nan"),
        "p_correction_p01_MPa": float(np.nanquantile(p_corr[valid], 0.01) / 1e6) if np.any(valid) else float("nan"),
        "p_correction_p99_MPa": float(np.nanquantile(p_corr[valid], 0.99) / 1e6) if np.any(valid) else float("nan"),
        "tmd_points_base": int(len(base_tmd)),
        "tmd_points_total": int(len(tmd_total)),
        "tmd_common_points": int(len(common)),
        "tmd_shift_mean_K": float(np.mean(shifts)) if shifts else float("nan"),
        "tmd_shift_min_K": float(np.min(shifts)) if shifts else float("nan"),
        "tmd_shift_max_K": float(np.max(shifts)) if shifts else float("nan"),
    }

    fields = {
        "Delta": Delta,
        "X": X,
        "a_assoc": a_assoc,
        "p_assoc": p_assoc,
        "P_total": P_total,
        "alpha_total": alpha_total,
        "kappa_total": kappa_total,
        "dP_drho_mol": dP_drho_mol,
        "valid": valid,
        "tmd_total": tmd_total,
    }

    return result, fields

# Conservative sweep.
# These ranges are intentionally modest. We want stable proof-of-concept first, not huge distortion.
K0_values = [1e-10, 3e-10, 1e-9, 3e-9]
epsK_values = [1000.0, 1400.0, 1800.0, 2200.0]
alpha_values = [0.05, 0.10, 0.25, 0.50]
lambda_values = [1e-4, 3e-4, 1e-3, 3e-3, 1e-2, 3e-2, 1e-1, 3e-1]

rows = []
field_cache = {}

for K0 in K0_values:
    for epsK in epsK_values:
        for alpha0 in alpha_values:
            for lam in lambda_values:
                result, fields = compute_trial(K0, epsK, alpha0, lam)
                rows.append(result)

                key = (K0, epsK, alpha0, lam)
                field_cache[key] = fields

csv_path = OUT / "association_sweep.csv"

with open(csv_path, "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=list(rows[0].keys()))
    writer.writeheader()
    writer.writerows(rows)

# Candidate filter:
# - no mechanical instability
# - not an absurd pressure correction
# - enough shared TMD points to compare
stable = [
    r for r in rows
    if r["bad_mechanical_frac_valid"] == 0.0
    and r["tmd_common_points"] >= 5
    and abs(r["p_correction_p99_MPa"]) < 100.0
    and abs(r["p_correction_p01_MPa"]) < 100.0
]

# Prefer candidates that shift TMD non-trivially but keep correction moderate.
def score_candidate(r):
    shift = abs(r["tmd_shift_mean_K"]) if np.isfinite(r["tmd_shift_mean_K"]) else 0.0
    pressure_penalty = abs(r["p_correction_p99_MPa"]) + abs(r["p_correction_p01_MPa"])
    return shift - 0.01 * pressure_penalty

stable_sorted = sorted(stable, key=score_candidate, reverse=True)
top = stable_sorted[:10]

if not top:
    print("No stable candidates found under this filter. The sweep still saved all rows.")
    chosen = rows[0]
else:
    chosen = top[0]

chosen_key = (
    chosen["K0_m3_per_mol"],
    chosen["epsK_K"],
    chosen["alpha0"],
    chosen["lambda"],
)

chosen_fields = field_cache[chosen_key]

summary = {
    "base_file": str(base_npz_path),
    "metric_file": str(metric_npz_path),
    "base_tmd_points": len(base_tmd),
    "n_grid_min": float(np.nanmin(n_grid)),
    "n_grid_max": float(np.nanmax(n_grid)),
    "n_grid_median": float(np.nanmedian(n_grid)),
    "sweep_count": len(rows),
    "stable_candidate_count": len(stable),
    "top_candidates": top,
    "chosen_candidate": chosen,
}

with open(OUT / "association_sweep_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

np.savez_compressed(
    OUT / "association_chosen_fields.npz",
    Ts=Ts,
    rho_mass=rho_mass,
    rho_mol=rho_mol,
    n_grid=n_grid,
    P_base=P_base,
    alpha_base=alpha_base,
    kappa_base=kappa_base,
    invalid_base=invalid_base,
    Delta=chosen_fields["Delta"],
    X=chosen_fields["X"],
    a_assoc=chosen_fields["a_assoc"],
    p_assoc=chosen_fields["p_assoc"],
    P_total=chosen_fields["P_total"],
    alpha_total=chosen_fields["alpha_total"],
    kappa_total=chosen_fields["kappa_total"],
    dP_drho_mol=chosen_fields["dP_drho_mol"],
    valid=chosen_fields["valid"],
)

def heatmap(Z, title, fname, cbar_label=""):
    plt.figure(figsize=(8, 5))
    cmap = plt.cm.viridis.copy()
    cmap.set_bad(color="lightgray")

    plt.imshow(
        np.ma.masked_invalid(Z),
        origin="lower",
        aspect="auto",
        extent=[float(rho_mass[0]), float(rho_mass[-1]), float(Ts[0]), float(Ts[-1])],
        cmap=cmap,
    )

    cb = plt.colorbar()
    if cbar_label:
        cb.set_label(cbar_label)

    plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
    plt.ylabel("T [K]")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(OUT / fname, dpi=180)
    plt.close()

valid_chosen = chosen_fields["valid"]

heatmap(n_grid, "Interpolated direct-metric network factor n(T,rho)", "heatmap_n_grid.png")
heatmap(np.where(valid_chosen, chosen_fields["p_assoc"] * chosen["lambda"] / 1e6, np.nan),
        "Chosen CDA pressure correction", "heatmap_pressure_correction_MPa.png", "MPa")
heatmap(np.where(valid_chosen, chosen_fields["alpha_total"], np.nan),
        "Chosen CDA alpha_p from total pressure", "heatmap_alpha_total.png", "1/K")
heatmap(np.where(valid_chosen, chosen_fields["dP_drho_mol"], np.nan),
        "Chosen CDA dP/drho_molar", "heatmap_dP_drho_molar.png", "Pa m^3/mol")
heatmap(np.where(valid_chosen, chosen_fields["kappa_total"], np.nan),
        "Chosen CDA kappa_T", "heatmap_kappa_total.png", "1/Pa")
heatmap(np.where(valid_chosen, chosen_fields["X"], np.nan),
        "Chosen unbonded-site fraction X", "heatmap_X.png")

# TMD comparison plot
base_tmd_rhos = np.array(sorted(base_tmd.keys()))
base_tmd_Ts = np.array([base_tmd[r] for r in base_tmd_rhos])

chosen_tmd = chosen_fields["tmd_total"]
chosen_tmd_rhos = np.array(sorted(chosen_tmd.keys()))
chosen_tmd_Ts = np.array([chosen_tmd[r] for r in chosen_tmd_rhos])

plt.figure(figsize=(7, 5))
if len(base_tmd_rhos):
    plt.plot(base_tmd_rhos, base_tmd_Ts, label="Base alpha_p=0")
if len(chosen_tmd_rhos):
    plt.plot(chosen_tmd_rhos, chosen_tmd_Ts, label="CDA corrected alpha_p=0")
plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
plt.ylabel("TMD diagnostic T [K]")
plt.title("TMD diagnostic comparison")
plt.legend()
plt.tight_layout()
plt.savefig(OUT / "plot_tmd_comparison.png", dpi=180)
plt.close()

print(json.dumps(summary, indent=2))

zip_name = "association_sweep_bundle.zip"

with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for p in OUT.rglob("*"):
        z.write(p, p)

print("Created:", zip_name)
```

```python
# CDA-EOS association correction sweep
# Run this after the base CoolProp run and the direct-metric curvature run.
# It uses:
# - base CoolProp fields
# - direct-metric n(T,rho)
# - SAFT-style a_assoc
# - pressure correction from the same free-energy term

import json, csv, zipfile, math
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

try:
    from scipy.interpolate import griddata
    from scipy.ndimage import gaussian_filter
except Exception:
    !pip install -q scipy
    from scipy.interpolate import griddata
    from scipy.ndimage import gaussian_filter

R_GAS = 8.31446261815324
M_WATER = 0.018015268
M_SITES = 4

OUT = Path("outputs_assoc_sweep")
OUT.mkdir(exist_ok=True)

def find_file(filename):
    hits = list(Path("/content").rglob(filename))
    if not hits:
        raise FileNotFoundError(
            f"Could not find {filename} anywhere under /content. "
            "You need to rerun or unzip the output that created it."
        )
    print(f"Found {filename}: {hits[0]}")
    return hits[0]

base_npz_path = find_file("base_fields_grid.npz")
metric_npz_path = find_file("metric_curvature_fields.npz")

print("Using base file:", base_npz_path)
print("Using metric curvature file:", metric_npz_path)

base = np.load(base_npz_path)
metric = np.load(metric_npz_path)

Ts = base["Ts"]
rho_mass = base["rhos"]
rho_mol = rho_mass / M_WATER

P_base = base["p"]
alpha_base = base["alpha_p"]
kappa_base = base["kappa_T"]
cp_base = base["cp"]
invalid_base = base["invalid"].astype(bool)

T_metric = metric["Ts"]
rho_metric = metric["rho_mass"]
n_metric = metric["n"]
R_metric = metric["R"]

valid_n = np.isfinite(n_metric)

TT, RR = np.meshgrid(Ts, rho_mass, indexing="ij")

points = []
vals = []

for i, T in enumerate(T_metric):
    for j, rho in enumerate(rho_metric):
        if valid_n[i, j]:
            points.append((float(T), float(rho)))
            vals.append(float(n_metric[i, j]))

points = np.array(points)
vals = np.array(vals)

if len(vals) < 3:
    raise RuntimeError("Not enough valid n(T,rho) points from metric curvature output.")

n_interp = griddata(points, vals, (TT, RR), method="linear")
n_near = griddata(points, vals, (TT, RR), method="nearest")
n_interp = np.where(np.isfinite(n_interp), n_interp, n_near)
n_interp = np.clip(n_interp, 0.0, 1.0)

# Smooth lightly so finite derivatives do not chase pixel-level interpolation noise.
n_grid = gaussian_filter(n_interp, sigma=1.5)
n_grid = np.clip(n_grid, 0.0, 1.0)

def tmd_from_alpha(alpha, invalid_mask):
    out = {}
    for j, rho in enumerate(rho_mass):
        a = alpha[:, j]
        bad = invalid_mask[:, j] | ~np.isfinite(a)

        for i in range(len(Ts) - 1):
            if bad[i] or bad[i + 1]:
                continue

            a0 = a[i]
            a1 = a[i + 1]

            if a0 == 0:
                out[float(rho)] = float(Ts[i])
                break

            if a0 * a1 < 0:
                T0 = Ts[i]
                T1 = Ts[i + 1]
                Tcross = T0 - a0 * (T1 - T0) / (a1 - a0)
                out[float(rho)] = float(Tcross)
                break

    return out

base_tmd = tmd_from_alpha(alpha_base, invalid_base)

def compute_trial(K0, epsK, alpha0, lam):
    """
    K0: effective bonding volume [m^3/mol]
    epsK: epsilon/R in kelvin, so exp(epsilon/(RT)) = exp(epsK/T)
    alpha0: baseline association factor
    lam: CDA coupling strength
    """
    Tmat, rhomat = np.meshgrid(Ts, rho_mol, indexing="ij")

    Delta0 = K0 * (np.exp(epsK / Tmat) - 1.0)
    Delta = Delta0 * (alpha0 + (1.0 - alpha0) * n_grid)

    A = rhomat * M_SITES * Delta

    # Stable small-A handling.
    X = np.where(
        A < 1e-12,
        1.0 - A,
        (-1.0 + np.sqrt(1.0 + 4.0 * A)) / (2.0 * A),
    )

    X = np.clip(X, 1e-12, 1.0)

    a_assoc = R_GAS * Tmat * M_SITES * (np.log(X) - X / 2.0 + 0.5)

    da_drho = np.gradient(a_assoc, rho_mol, axis=1, edge_order=2)
    p_assoc = (rho_mol[None, :] ** 2) * da_drho

    P_total = P_base + lam * p_assoc

    dP_dT = np.gradient(P_total, Ts, axis=0, edge_order=2)
    dP_drho_mol = np.gradient(P_total, rho_mol, axis=1, edge_order=2)

    alpha_total = (1.0 / rho_mol[None, :]) * dP_dT / dP_drho_mol
    kappa_total = 1.0 / (rho_mol[None, :] * dP_drho_mol)

    valid = (
        (~invalid_base)
        & np.isfinite(P_total)
        & np.isfinite(alpha_total)
        & np.isfinite(kappa_total)
        & np.isfinite(dP_drho_mol)
    )

    denom = max(int(np.sum(valid)), 1)

    bad_mech = ((dP_drho_mol <= 0.0) | (kappa_total <= 0.0)) & valid

    tmd_total = tmd_from_alpha(alpha_total, invalid_base | (~valid))

    common = sorted(set(base_tmd.keys()).intersection(tmd_total.keys()))
    shifts = [tmd_total[r] - base_tmd[r] for r in common]

    p_corr = lam * p_assoc

    result = {
        "K0_m3_per_mol": float(K0),
        "epsK_K": float(epsK),
        "alpha0": float(alpha0),
        "lambda": float(lam),
        "valid_cells": int(np.sum(valid)),
        "bad_mechanical_cells": int(np.sum(bad_mech)),
        "bad_mechanical_frac_valid": float(np.sum(bad_mech) / denom),
        "min_dP_drho_molar_valid": float(np.nanmin(dP_drho_mol[valid])) if np.any(valid) else float("nan"),
        "min_kappa_total_valid": float(np.nanmin(kappa_total[valid])) if np.any(valid) else float("nan"),
        "p_correction_median_MPa": float(np.nanmedian(p_corr[valid]) / 1e6) if np.any(valid) else float("nan"),
        "p_correction_p01_MPa": float(np.nanquantile(p_corr[valid], 0.01) / 1e6) if np.any(valid) else float("nan"),
        "p_correction_p99_MPa": float(np.nanquantile(p_corr[valid], 0.99) / 1e6) if np.any(valid) else float("nan"),
        "tmd_points_base": int(len(base_tmd)),
        "tmd_points_total": int(len(tmd_total)),
        "tmd_common_points": int(len(common)),
        "tmd_shift_mean_K": float(np.mean(shifts)) if shifts else float("nan"),
        "tmd_shift_min_K": float(np.min(shifts)) if shifts else float("nan"),
        "tmd_shift_max_K": float(np.max(shifts)) if shifts else float("nan"),
    }

    fields = {
        "Delta": Delta,
        "X": X,
        "a_assoc": a_assoc,
        "p_assoc": p_assoc,
        "P_total": P_total,
        "alpha_total": alpha_total,
        "kappa_total": kappa_total,
        "dP_drho_mol": dP_drho_mol,
        "valid": valid,
        "tmd_total": tmd_total,
    }

    return result, fields

# Conservative sweep.
# These ranges are intentionally modest. We want stable proof-of-concept first, not huge distortion.
K0_values = [1e-10, 3e-10, 1e-9, 3e-9]
epsK_values = [1000.0, 1400.0, 1800.0, 2200.0]
alpha_values = [0.05, 0.10, 0.25, 0.50]
lambda_values = [1e-4, 3e-4, 1e-3, 3e-3, 1e-2, 3e-2, 1e-1, 3e-1]

rows = []
field_cache = {}

for K0 in K0_values:
    for epsK in epsK_values:
        for alpha0 in alpha_values:
            for lam in lambda_values:
                result, fields = compute_trial(K0, epsK, alpha0, lam)
                rows.append(result)

                key = (K0, epsK, alpha0, lam)
                field_cache[key] = fields

csv_path = OUT / "association_sweep.csv"

with open(csv_path, "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=list(rows[0].keys()))
    writer.writeheader()
    writer.writerows(rows)

# Candidate filter:
# - no mechanical instability
# - not an absurd pressure correction
# - enough shared TMD points to compare
stable = [
    r for r in rows
    if r["bad_mechanical_frac_valid"] == 0.0
    and r["tmd_common_points"] >= 5
    and abs(r["p_correction_p99_MPa"]) < 100.0
    and abs(r["p_correction_p01_MPa"]) < 100.0
]

# Prefer candidates that shift TMD non-trivially but keep correction moderate.
def score_candidate(r):
    shift = abs(r["tmd_shift_mean_K"]) if np.isfinite(r["tmd_shift_mean_K"]) else 0.0
    pressure_penalty = abs(r["p_correction_p99_MPa"]) + abs(r["p_correction_p01_MPa"])
    return shift - 0.01 * pressure_penalty

stable_sorted = sorted(stable, key=score_candidate, reverse=True)
top = stable_sorted[:10]

if not top:
    print("No stable candidates found under this filter. The sweep still saved all rows.")
    chosen = rows[0]
else:
    chosen = top[0]

chosen_key = (
    chosen["K0_m3_per_mol"],
    chosen["epsK_K"],
    chosen["alpha0"],
    chosen["lambda"],
)

chosen_fields = field_cache[chosen_key]

summary = {
    "base_file": str(base_npz_path),
    "metric_file": str(metric_npz_path),
    "base_tmd_points": len(base_tmd),
    "n_grid_min": float(np.nanmin(n_grid)),
    "n_grid_max": float(np.nanmax(n_grid)),
    "n_grid_median": float(np.nanmedian(n_grid)),
    "sweep_count": len(rows),
    "stable_candidate_count": len(stable),
    "top_candidates": top,
    "chosen_candidate": chosen,
}

with open(OUT / "association_sweep_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

np.savez_compressed(
    OUT / "association_chosen_fields.npz",
    Ts=Ts,
    rho_mass=rho_mass,
    rho_mol=rho_mol,
    n_grid=n_grid,
    P_base=P_base,
    alpha_base=alpha_base,
    kappa_base=kappa_base,
    invalid_base=invalid_base,
    Delta=chosen_fields["Delta"],
    X=chosen_fields["X"],
    a_assoc=chosen_fields["a_assoc"],
    p_assoc=chosen_fields["p_assoc"],
    P_total=chosen_fields["P_total"],
    alpha_total=chosen_fields["alpha_total"],
    kappa_total=chosen_fields["kappa_total"],
    dP_drho_mol=chosen_fields["dP_drho_mol"],
    valid=chosen_fields["valid"],
)

def heatmap(Z, title, fname, cbar_label=""):
    plt.figure(figsize=(8, 5))
    cmap = plt.cm.viridis.copy()
    cmap.set_bad(color="lightgray")

    plt.imshow(
        np.ma.masked_invalid(Z),
        origin="lower",
        aspect="auto",
        extent=[float(rho_mass[0]), float(rho_mass[-1]), float(Ts[0]), float(Ts[-1])],
        cmap=cmap,
    )

    cb = plt.colorbar()
    if cbar_label:
        cb.set_label(cbar_label)

    plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
    plt.ylabel("T [K]")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(OUT / fname, dpi=180)
    plt.close()

valid_chosen = chosen_fields["valid"]

heatmap(n_grid, "Interpolated direct-metric network factor n(T,rho)", "heatmap_n_grid.png")
heatmap(np.where(valid_chosen, chosen_fields["p_assoc"] * chosen["lambda"] / 1e6, np.nan),
        "Chosen CDA pressure correction", "heatmap_pressure_correction_MPa.png", "MPa")
heatmap(np.where(valid_chosen, chosen_fields["alpha_total"], np.nan),
        "Chosen CDA alpha_p from total pressure", "heatmap_alpha_total.png", "1/K")
heatmap(np.where(valid_chosen, chosen_fields["dP_drho_mol"], np.nan),
        "Chosen CDA dP/drho_molar", "heatmap_dP_drho_molar.png", "Pa m^3/mol")
heatmap(np.where(valid_chosen, chosen_fields["kappa_total"], np.nan),
        "Chosen CDA kappa_T", "heatmap_kappa_total.png", "1/Pa")
heatmap(np.where(valid_chosen, chosen_fields["X"], np.nan),
        "Chosen unbonded-site fraction X", "heatmap_X.png")

# TMD comparison plot
base_tmd_rhos = np.array(sorted(base_tmd.keys()))
base_tmd_Ts = np.array([base_tmd[r] for r in base_tmd_rhos])

chosen_tmd = chosen_fields["tmd_total"]
chosen_tmd_rhos = np.array(sorted(chosen_tmd.keys()))
chosen_tmd_Ts = np.array([chosen_tmd[r] for r in chosen_tmd_rhos])

plt.figure(figsize=(7, 5))
if len(base_tmd_rhos):
    plt.plot(base_tmd_rhos, base_tmd_Ts, label="Base alpha_p=0")
if len(chosen_tmd_rhos):
    plt.plot(chosen_tmd_rhos, chosen_tmd_Ts, label="CDA corrected alpha_p=0")
plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
plt.ylabel("TMD diagnostic T [K]")
plt.title("TMD diagnostic comparison")
plt.legend()
plt.tight_layout()
plt.savefig(OUT / "plot_tmd_comparison.png", dpi=180)
plt.close()

print(json.dumps(summary, indent=2))

zip_name = "association_sweep_bundle.zip"

with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for p in OUT.rglob("*"):
        z.write(p, p)

print("Created:", zip_name)
```

```python
# CDA-EOS Cp validation prototype
# Run this after association_sweep_bundle was generated.
# It computes Cp_total from the fully assembled pressure/free-energy correction
# using the thermodynamic identity:
# Cp = (dh/dT)_rho - (dh/drho)_T * (dp/dT)_rho / (dp/drho)_T

import json, zipfile, csv
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

R_GAS = 8.31446261815324
M_WATER = 0.018015268

OUT = Path("outputs_cp_validation")
OUT.mkdir(exist_ok=True)

def find_all(filename):
    return list(Path("/content").rglob(filename))

def pick_file(filename):
    hits = find_all(filename)
    if not hits:
        raise FileNotFoundError(f"Could not find {filename} anywhere under /content.")
    print(f"Found {filename}: {hits[0]}")
    return hits[0]

chosen_path = pick_file("association_chosen_fields.npz")
summary_path = pick_file("association_sweep_summary.json")

chosen = np.load(chosen_path)

Ts = chosen["Ts"]
rho_mass = chosen["rho_mass"]
rho_mol = chosen["rho_mol"]

P_base = chosen["P_base"]
P_total = chosen["P_total"]
p_assoc = chosen["p_assoc"]
a_assoc = chosen["a_assoc"]

alpha_base = chosen["alpha_base"]
kappa_base = chosen["kappa_base"]
invalid_base = chosen["invalid_base"].astype(bool)
valid_assoc = chosen["valid"].astype(bool)

with open(summary_path) as f:
    sweep_summary = json.load(f)

chosen_candidate = sweep_summary["chosen_candidate"]
lam = float(chosen_candidate["lambda"])

# Find a matching base_fields_grid.npz, because we need base internal energy u.
base_candidates = find_all("base_fields_grid.npz")
if not base_candidates:
    raise FileNotFoundError("Could not find base_fields_grid.npz. Rerun the base CoolProp proof.")

base = None
base_path = None

for p in base_candidates:
    b = np.load(p)
    if (
        "Ts" in b.files
        and "rhos" in b.files
        and "u" in b.files
        and "cp" in b.files
        and b["Ts"].shape == Ts.shape
        and b["rhos"].shape == rho_mass.shape
        and np.allclose(b["Ts"], Ts)
        and np.allclose(b["rhos"], rho_mass)
    ):
        base = b
        base_path = p
        break

if base is None:
    raise RuntimeError("Found base_fields_grid.npz files, but none matched the association grid.")

print("Using matched base file:", base_path)

u_base_mass = base["u"]       # J/kg
cp_base_mass = base["cp"]     # J/(kg K)
P_base_check = base["p"]      # Pa

# Molar enthalpy base:
# h = u + p/rho_mass on mass basis, then convert to molar.
rho_mass_grid = rho_mass[None, :]
rho_mol_grid = rho_mol[None, :]
T_grid = Ts[:, None]

h_base_mol = (u_base_mass + P_base_check / rho_mass_grid) * M_WATER

# Association entropy and enthalpy correction.
# s_assoc = -(partial a_assoc / partial T)_rho
da_assoc_dT = np.gradient(a_assoc, Ts, axis=0, edge_order=2)
s_assoc = -da_assoc_dT

# h_corr = lambda * (a_assoc + T s_assoc + p_assoc/rho)
h_corr_mol = lam * (a_assoc + T_grid * s_assoc + p_assoc / rho_mol_grid)

h_total_mol = h_base_mol + h_corr_mol

# Compute Cp_total from:
# Cp = h_T - h_rho * p_T / p_rho
h_T = np.gradient(h_total_mol, Ts, axis=0, edge_order=2)
h_rho = np.gradient(h_total_mol, rho_mol, axis=1, edge_order=2)

p_T = np.gradient(P_total, Ts, axis=0, edge_order=2)
p_rho = np.gradient(P_total, rho_mol, axis=1, edge_order=2)

Cp_total_mol = h_T - h_rho * (p_T / p_rho)
Cp_total_mass = Cp_total_mol / M_WATER

# Base identity check:
# If this is bad, the grid derivative is too rough.
h_base_T = np.gradient(h_base_mol, Ts, axis=0, edge_order=2)
h_base_rho = np.gradient(h_base_mol, rho_mol, axis=1, edge_order=2)

p_base_T = np.gradient(P_base_check, Ts, axis=0, edge_order=2)
p_base_rho = np.gradient(P_base_check, rho_mol, axis=1, edge_order=2)

Cp_base_identity_mol = h_base_T - h_base_rho * (p_base_T / p_base_rho)
Cp_base_identity_mass = Cp_base_identity_mol / M_WATER

# Robust mask:
# remove invalid, nonfinite, boundary, and cells touching invalid regions.
robust = (
    valid_assoc
    & (~invalid_base)
    & np.isfinite(Cp_total_mass)
    & np.isfinite(cp_base_mass)
    & np.isfinite(Cp_base_identity_mass)
    & np.isfinite(p_rho)
    & (p_rho > 0)
)

robust[0, :] = False
robust[-1, :] = False
robust[:, 0] = False
robust[:, -1] = False

bad_seed = invalid_base | (~valid_assoc) | (~np.isfinite(P_total))

for di in [-1, 0, 1]:
    for dj in [-1, 0, 1]:
        shifted = np.roll(np.roll(bad_seed, di, axis=0), dj, axis=1)

        if di > 0:
            shifted[:di, :] = False
        if di < 0:
            shifted[di:, :] = False
        if dj > 0:
            shifted[:, :dj] = False
        if dj < 0:
            shifted[:, dj:] = False

        robust &= ~shifted

Cp_delta_mass = Cp_total_mass - cp_base_mass
Cp_ratio = Cp_total_mass / cp_base_mass

base_identity_rel_error = (Cp_base_identity_mass - cp_base_mass) / cp_base_mass
Cp_rel_change = Cp_delta_mass / cp_base_mass

def stats(arr, mask):
    vals = arr[mask]
    vals = vals[np.isfinite(vals)]
    if vals.size == 0:
        return {
            "min": float("nan"),
            "p01": float("nan"),
            "median": float("nan"),
            "p99": float("nan"),
            "max": float("nan"),
        }
    return {
        "min": float(np.nanmin(vals)),
        "p01": float(np.nanquantile(vals, 0.01)),
        "median": float(np.nanmedian(vals)),
        "p99": float(np.nanquantile(vals, 0.99)),
        "max": float(np.nanmax(vals)),
    }

report = {
    "method": "Cp prototype from h_total and p_total finite derivatives",
    "chosen_candidate": chosen_candidate,
    "base_file": str(base_path),
    "association_file": str(chosen_path),
    "total_grid_cells": int(Cp_total_mass.size),
    "robust_cp_cells": int(np.sum(robust)),
    "robust_cp_fraction": float(np.mean(robust)),
    "Cp_total_mass_stats_J_per_kgK": stats(Cp_total_mass, robust),
    "Cp_base_mass_stats_J_per_kgK": stats(cp_base_mass, robust),
    "Cp_delta_mass_stats_J_per_kgK": stats(Cp_delta_mass, robust),
    "Cp_ratio_stats": stats(Cp_ratio, robust),
    "Cp_relative_change_stats": stats(Cp_rel_change, robust),
    "base_Cp_identity_relative_error_stats": stats(base_identity_rel_error, robust),
    "Cp_total_le_0_cells": int(np.sum((Cp_total_mass <= 0) & robust)),
    "Cp_total_le_0_frac_robust": float(
        np.sum((Cp_total_mass <= 0) & robust) / max(int(np.sum(robust)), 1)
    ),
    "p_rho_le_0_cells": int(np.sum((p_rho <= 0) & robust)),
    "p_rho_le_0_frac_robust": float(
        np.sum((p_rho <= 0) & robust) / max(int(np.sum(robust)), 1)
    ),
}

with open(OUT / "cp_validation_report.json", "w") as f:
    json.dump(report, f, indent=2)

np.savez_compressed(
    OUT / "cp_validation_fields.npz",
    Ts=Ts,
    rho_mass=rho_mass,
    rho_mol=rho_mol,
    robust=robust,
    Cp_total_mass=Cp_total_mass,
    cp_base_mass=cp_base_mass,
    Cp_base_identity_mass=Cp_base_identity_mass,
    Cp_delta_mass=Cp_delta_mass,
    Cp_ratio=Cp_ratio,
    Cp_rel_change=Cp_rel_change,
    base_identity_rel_error=base_identity_rel_error,
    h_base_mol=h_base_mol,
    h_corr_mol=h_corr_mol,
    h_total_mol=h_total_mol,
    p_rho=p_rho,
    p_T=p_T,
)

def heatmap(Z, title, fname, cbar_label=""):
    plt.figure(figsize=(8, 5))
    cmap = plt.cm.viridis.copy()
    cmap.set_bad(color="lightgray")

    plt.imshow(
        np.ma.masked_invalid(np.where(robust, Z, np.nan)),
        origin="lower",
        aspect="auto",
        extent=[float(rho_mass[0]), float(rho_mass[-1]), float(Ts[0]), float(Ts[-1])],
        cmap=cmap,
    )

    cb = plt.colorbar()
    if cbar_label:
        cb.set_label(cbar_label)

    plt.xlabel(r"$\rho_{\rm mass}$ [kg/m$^3$]")
    plt.ylabel("T [K]")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(OUT / fname, dpi=180)
    plt.close()

heatmap(Cp_total_mass, "CDA prototype Cp_total", "heatmap_Cp_total.png", "J/(kg K)")
heatmap(Cp_delta_mass, "CDA prototype Cp_total - Cp_base", "heatmap_Cp_delta.png", "J/(kg K)")
heatmap(Cp_ratio, "CDA prototype Cp_total / Cp_base", "heatmap_Cp_ratio.png")
heatmap(Cp_rel_change, "Relative Cp change", "heatmap_Cp_relative_change.png")
heatmap(base_identity_rel_error, "Base Cp identity relative error", "heatmap_base_Cp_identity_error.png")
heatmap(h_corr_mol, "Association enthalpy correction", "heatmap_h_corr_mol.png", "J/mol")
heatmap(np.where(robust, 1.0, np.nan), "Robust Cp validation cells", "heatmap_cp_robust_mask.png")

# Line plot near rho_mass = 1000 kg/m^3
j1000 = int(np.argmin(np.abs(rho_mass - 1000.0)))

plt.figure(figsize=(7, 5))
plt.plot(Ts, cp_base_mass[:, j1000], label="Base Cp")
plt.plot(Ts, Cp_total_mass[:, j1000], label="CDA Cp prototype")
plt.xlabel("T [K]")
plt.ylabel("Cp [J/(kg K)]")
plt.title(f"Cp comparison near rho_mass = {rho_mass[j1000]:.1f} kg/m^3")
plt.legend()
plt.tight_layout()
plt.savefig(OUT / "plot_Cp_comparison_rho1000.png", dpi=180)
plt.close()

plt.figure(figsize=(7, 5))
plt.plot(Ts, Cp_delta_mass[:, j1000])
plt.axhline(0.0)
plt.xlabel("T [K]")
plt.ylabel("Cp_total - Cp_base [J/(kg K)]")
plt.title(f"Cp correction near rho_mass = {rho_mass[j1000]:.1f} kg/m^3")
plt.tight_layout()
plt.savefig(OUT / "plot_Cp_delta_rho1000.png", dpi=180)
plt.close()

print(json.dumps(report, indent=2))

zip_name = "cp_validation_bundle.zip"

with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for p in OUT.rglob("*"):
        z.write(p, p)

print("Created:", zip_name)
```
